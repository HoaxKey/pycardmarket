```python
import pandas as pd
```

```python
export_df = pd.read_csv('products_export_example.csv', sep = ',')
export_df.head()
```

```python
expansions = ['Arcane Rising', 'Crucible of War', 'Dynasty', 'Everfest', 'Monarch', 'Tales of Aria', 'Uprising', 'Welcome to Rathe', 'Promos', 'History Pack 1']
to_buy = []
handles = export_df['Handle'].unique()
for handle in handles:
    handle_df = export_df[export_df['Handle'] == handle]
    buying_df = handle_df[handle_df['Cost per item'].notnull()]
    if len(buying_df.index) >0 :
        tags = ''
        handle_title = ''
        for index, row in handle_df.iterrows():
            if row['Tags'] != '':
                tags =  row['Tags']
                handle_title = row['Title']
                break
        handle_expansion = ''
        for expansion in expansions:
            if expansion in tags:
                handle_expansion = expansion
                break
                
    
        for index, row in buying_df.iterrows():
            #print(f"{handle_title}, {row['Option1 Value']}, {row['Option2 Value']}")
            try:
                if row['Google Shopping / Custom Label 0'] > row['Variant Inventory Qty']:
                    #We should buy this item
                    to_buy.append([handle_expansion, handle_title, row['Option1 Value'], row['Option2 Value'], row['Cost per item'], int(row['Google Shopping / Custom Label 0'] - row['Variant Inventory Qty'])])
            except KeyError:
                print(row)
    
    

to_buy_df = pd.DataFrame(to_buy, columns = ['Expansion', 'Title', 'Variant', 'Edition', 'Price', 'Number'])
to_buy_df.head(100)
    
```

```python
expansions = ['Arcane Rising', 'Crucible of War', 'Dynasty', 'Everfest', 'Monarch', 'Tales of Aria', 'Uprising', 'Welcome to Rathe', 'Promos', 'History Pack 1']
for expansion in expansions:
    print(f'\n{expansion}\n')
    expansion_df = to_buy_df[to_buy_df['Expansion'] == expansion]
    df_sorted = expansion_df.sort_values('Title')
    for index, row in df_sorted.iterrows():
        print(f"{row['Title']}, {row['Variant']}, {row['Edition']}, £{'%.2f' % (row['Price'])}, {row['Number']}")
    
```
